### Null Values

```python
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
```

```python
last_ten_year=df[["type","release_year"]]
last_ten_year=last_ten_year[last_ten_year["release_year"] >=2012]
last_ten_year
```

```python
last_ten_year.groupby("release_year")["type"].value_counts()
```

```python
last_ten_year_df =last_ten_year.groupby("release_year")["type"].size().reset_index()
last_ten_year_df = pd.DataFrame(last_ten_year_df)
last_ten_year_df
```

```python
value_ranges = {}
for column in df.columns:
    min_value = df[column].min()
    max_value = df[column].max()
    value_ranges[column] = (min_value, max_value)
print(value_ranges)
```

```python
from scipy.stats import gamma
data_gamma = gamma.rvs(a=5, size=1000)
sns.histplot(data_gamma, bins=100, kde=True, color = 'orange', element='step', fill=True)
plt.xlabel('Gamma Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import norm

data_normal = norm.rvs(size=1000,loc=0,scale=1)
sns.histplot(data_normal, bins=100, kde=True, color='green', element='step', fill=True)
plt.xlabel('Normal Distribution')
plt.ylabel('Frequency')
plt.show()
```
